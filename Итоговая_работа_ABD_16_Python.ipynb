**4. Добавим к отфильтрованному по наиболее активному пользователю df необходимые столбцы: жанры, общее количество оценок на фильм от всех пользователей, суммарная оценка на фильм от всех пользователей**

```python
super_user.head(5)
```

```python
movies.head(5)
```

```python
super_user_genres=super_user.merge(movies,how='inner',left_on='id_film',right_on='id') #присоединям таблицу жанров
super_user_genres=super_user_genres.drop(columns=['id','empty','url imdb','timestamp'],axis=1)         # и уберем ненужные столбцы 
super_user_genres
```

```python
ratings.head(5)
```

```python
user_grade_ml=ratings.groupby(by='id_film').agg({'grade':'count'}) #количество оценок каждого фильма всеми пользователями
user_grade_ml = user_grade_ml.reset_index()
user_grade_ml = user_grade_ml.rename(columns={'grade':'count_grade'}) #переименуем, чтобы столбцы с оценками при дальнейшем соединении таблиц имели разные имена
user_grade_ml
```

```python
super_user_genres_count=super_user_genres.merge(user_grade_ml,how='inner',on='id_film') #присоединям количество оценок всех пользователей
super_user_genres_count
```

```python
super_user_genres_count.columns
```

```python
super_user_genres_count = super_user_genres_count.reindex(columns=['id_user', 'id_film', 'grade', 'count_grade', 'name', 'release date', 'unknown',
       'action', 'adventure', 'animation', 'children', 'comedy', 'criminal',
       'documentary', 'drama', 'fantasy', 'nuar', 'horror', 'music', 'mystic',
       'romance', 'fiction', 'thriller', 'war', 'western']) #для удобства переместим колонку count_grade рядом с grade
super_user_genres_count       
```

```python
user_grade_ml_1=ratings.groupby(by='id_film').agg({'grade':'sum'}) #сумма оценок каждого фильма всеми пользователями
user_grade_ml_1 = user_grade_ml_1.reset_index()
user_grade_ml_1 = user_grade_ml_1.rename(columns={'grade':'sum_grade'}) #переименуем, чтобы столбцы с оценками при дальнейшем соединении таблиц имели разные имена
user_grade_ml_1
```

```python
super_user_genres_count_sum=super_user_genres_count.merge(user_grade_ml_1,how='inner',on='id_film') #присоединям сумму оценок всех пользователей
super_user_genres_count_sum
```

```python
super_user_genres_count_sum.columns
```

```python
super_user_genres_count_sum = super_user_genres_count_sum.reindex(columns=['id_user', 'id_film', 'grade', 'count_grade', 'sum_grade', 'name', 'release date',
       'unknown', 'action', 'adventure', 'animation', 'children', 'comedy',
       'criminal', 'documentary', 'drama', 'fantasy', 'nuar', 'horror',
       'music', 'mystic', 'romance', 'fiction', 'thriller', 'war', 'western'
       ]) #для удобства переместим колонку sum_grade рядом с count_grade
super_user_genres_count_sum
```

```python
#для дальнейшей работы назовем полученный df более компактно и заменим значения строкового столбца release date на числовые (год)
import datetime as dt
recom_model=super_user_genres_count_sum
recom_model['release date'] = pd.to_datetime(recom_model['release date'])
recom_model['release date'] = recom_model['release date'].dt.year
```

```python
recom_model
```

Теперь все готово и можно строить модель


**5. Сформируем X_train, X_test, y_train, y_test**

```python
#Признаки:год выхода, жанры,количество оценок каждого фильма всеми пользователями,суммарная оценка каждого фильма всеми пользователями
#Целевая переменная для рекомендации-оценка фильма пользователем id=405
X,y=recom_model[['release date','unknown', 'action', 'adventure', 'animation', 'children', 'comedy',
       'criminal', 'documentary', 'drama', 'fantasy', 'nuar', 'horror',
       'music', 'mystic', 'romance', 'fiction', 'thriller', 'war', 'western','count_grade','sum_grade']],recom_model['grade']
```

```python
X #наша таблица признаков
```

```python
y #правильные ответы на основании признаков
```

```python
#импортируем из библиотеки sklearn метод разделения выборки
from sklearn.model_selection import train_test_split
```

**6. Возьмем модель линейной регрессии и обучим ее на фильмах**

```python
#импортируем модель линейной регрессии из sklearn
from sklearn.linear_model import LinearRegression
```

```python
#создаем модель
model=LinearRegression()
```

```python
#обучаем модель на тренировочной выборке
model.fit(X_train,y_train)
```

**7. Оценим качество модели на X_test, y_test при помощи метрик для задачи регрессии**

```python
#Загружаем данные в spark
ratings_s=spark.read.csv('u.data(1).csv',inferSchema=True,header=True) #заголовки были даны в гугл-таблицах (импорт в таблицы--дали заголовки--скачивание в формате csv)  
ratings_s.show(5)
```

**9. Средствами спарка выведем среднюю оценку для каждого фильма**

```python
ratings_s=ratings_s.withColumnRenamed('id_film','id') #переименовала колонку для точного совпадения названия с колонкой присоединяемой таблицы
```

```python
ratings_s_avg= ratings_s.groupBy('id').avg('grade') #группировка оценок по фильмам с подсчетом средней оценки по каждому фильму
ratings_s_avg=ratings_s_avg.withColumnRenamed('avg(grade)','avg_grade') #переименование колонки
ratings_s_avg.show(5)
```

```python
movies_s1=movies_s.select('id','name') #из таблицы фильмов возьмем id (для соединения) и названия фильмов
movies_s1.show()
```

```python
ratings_s1_avg=movies_s1.join(ratings_s_avg,'id',how='inner') #присоединили к таблице фильмов их средние оценки 
ratings_s1_avg.show(5)
```

**10. Посчитаем средствами спарка среднюю оценку для каждого жанра**

```python
avg_grade_genre=ratings_s1_avg.join(movies_s,'id',how='inner') #соединим полученные средние оценки фильмов с таблицей фильмов по жанрам
avg_grade_genre.show() 
```

```python
avg_grade_genre= avg_grade_genre.select('avg_grade','unknown','action','adventure','animation',"children's",'comedy',
                                         'criminal','documentary','drama','fantasy','nuar','horror','music',
                                         'mystic','romance','fiction','thriller','war','western')
avg_grade_genre.show(5) #теперь мы имеем строки со средними оценками фильмов и жанры этих фильмов
```

```python
##сгруппируем колонку каждого жанра по значениям (их всего два-0 и 1) и посчитаем новое среднее значение.Делать будем в цикле, чтобы автоматизировать процесс и сократить код. 
for genre in ('unknown','action','adventure','animation',"children's",'comedy','criminal','documentary','drama','fantasy','nuar','horror','music','mystic','romance','fiction','thriller','war','western'):
  genres=avg_grade_genre.groupBy(genre).avg('avg_grade')
  genres=genres.where(genres[genre]==1) #нам нужна средняя оценка на фильмах, где значение жанра равно 1
  genres.show(5)
```

**11. В спарке получить 2 датафрейма с 5-ю самыми популярными и самыми непопулярными фильмами. Популярность договоримся считать по количеству оценок(т.к.количество оценок в принципе говорит о количестве просмотров)**

```python
popular=movies_s.select('id','name') #таблица фильмов с нужными колонками
popular.show(5)
```

```python
grades=ratings_s.select('id') #оцениваемые фильмы пользователями
grades=grades.groupBy('id').count() #как часто пользователи смотрели+оценивали каждый фильм
grades.show()
```

```python
popular=popular.join(grades,'id',how='inner') #получили df с количеством оценок каждого фильма и названиями этих фильмов
popular.show(5)
```

```python
#теперь необходимо отсортировать полученный df по возрастанию количества оценок, чтобы получить фильмы по убыванию популярности
top_popular=popular.sort('count', ascending=False)
top_popular.show(5)
```

```python
top_popular=top_popular.take(5) #возьмем первые 5 строк, это и есть самые популярные по количеству просмотров/оценок фильмы
```

```python
new=['id','name','count_grades']
```

```python
top_popular=pd.DataFrame(top_popular,columns=new)
```

```python
top_popular #Получили датафрейм с 5-ю самыми популярными фильмами.
```

```python
#теперь необходимо отсортировать полученный df по УБЫВАНИЮ количества оценок, чтобы получить фильмы по ВОЗРАСТАНИЮ популярности
top_unpopular=popular.sort('count', ascending=True)
top_unpopular.show(5)
```

```python
top_unpopular=top_unpopular.take(5) #возьмем первые 5 строк, это и есть самые НЕпопулярные по количеству просмотров/оценок фильмы
```

```python
top_unpopular=pd.DataFrame(top_unpopular,columns=new)
```
